```python
import numpy as np
import matplotlib.pyplot as plt

# N√∫mero de amostras
N = 100

# Matrizes definidas na imagem (Equa√ß√£o 2.54)
A1 = np.array([
    [0.4, 1.2],
    [0.3, 0.7]
])

A2 = np.array([
    [0.35, -0.3],
    [-0.4, -0.5]
])

# Matriz B (identidade na imagem)
B = np.eye(2)

# Entradas ex√≥genas u(k): ru√≠do branco gaussiano
# Criamos uma matriz (2 linhas, N colunas)
u = np.random.normal(0, 1, (2, N))

# Inicializa√ß√£o do vetor de sa√≠da y: (2 linhas, N colunas)
y = np.zeros((2, N))

# Loop de simula√ß√£o
# Come√ßamos em 2 para ter acesso a y(k-1) e y(k-2)
for k in range(2, N):
    # C√°lculo de cada termo da equa√ß√£o
    termo_y_k1 = A1 @ y[:, k-1]          # A1 * y(k-1)
    termo_y_k2 = A2 @ y[:, k-2]          # A2 * y(k-2)
    termo_u_k1 = B @ u[:, k-1]           # B * u(k-1)

    # Soma dos componentes para obter y(k)
    y[:, k] = termo_y_k1 + termo_y_k2 + termo_u_k1

# Plotagem dos resultados
plt.figure(figsize=(12, 6))
plt.plot(y[0, :], label='$y_1(k)$', color='blue')
plt.plot(y[1, :], label='$y_2(k)$', color='red')
plt.title('Simula√ß√£o do Modelo VARX (Eq. 2.54)')
plt.xlabel('Amostras (k)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
