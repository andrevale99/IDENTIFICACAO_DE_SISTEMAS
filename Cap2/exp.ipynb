### Aproximaçã de atraso puro de tempo

Uma das maneiras de aproximar o atraso no sistema é
utilizando a aproximação de Padé. Esse método aproxima
uma funçao qualquer por uma razão de polinômios em $s$.


Um atraso no domínio $s$ é representado por uma multiplicação de
$e^{st}$ no tempo, de acordo com as propriedades de Laplace:

$e^{s_0t}x(t) → X(s - s_0)$

Resolver a integral da transformada de Laplace se torna complexa, logo utiliza-se da aproximação de Padé:

$e^{-\tau_ds} ≈ R_n(s) = \frac{Q_n(-\tau_ds)}{Q_n(\tau_ds)}$

$Q_n(\tau_ds) = \sum_{j=0}^{n} \frac{(n+j)!}{j!(n-j)!}(\tau_ds)^{(n-j)}$

Usando um atraso de 1 unidade $(\tau_d = 1)$ a aproximação
com $R_1(s)$ ficará:

$Q_n(\tau_ds) = \sum_{j=0}^{n} \frac{(n+j)!}{j!(n-j)!}(\tau_ds)^{(n-j)}$

$Q_1(1s) = \frac{(1+0)!}{0!(1-0)!}(1s)^{(1-0)} + \frac{(1+1)!}{1!(1-1)!}(1s)^{(1-1)} = 2 + s$

De forma analoga

$Q_1(-1s) = -s + 2$

Portanto o polinômio $R_1(s)$ será:

$R_1(s) = \frac{Q_n(-1s)}{Q_n(1s)} = \frac{-s + 2}{s + 2}$


Por fim, substituirmos $e^{-\tau_ds}$ na equação da função de transferência.

$e^{-\tau_ds} = R_1(s)$

O exemplo do livro considera que:

$H(s) = \frac{e^{-\tau_ds}}{(s+1)}$

Como $\tau_d = 1$, logo:

$H(s) = \frac{e^{-1s}}{(s+1)}$

E:

$R_1(s) = \frac{-s+2}{s+2}$

Portanto:

$H(s) =  \frac{\frac{-s+2}{s+2}}{(s+1)} = \frac{-s+2}{(s+2)(s+1)} = \frac{-s+2}{s^2 + 3s + 2}$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal as sg

Ns = [-1, 2]
Ds = [1, 3, 2]

Hs = sg.TransferFunction(Ns,Ds)

t = np.linspace(0, 5, 1000)
tout, yout2 = sg.step(Hs, T=t)

plt.plot(tout, yout2, color='red')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída')
plt.title('Resposta ao degrau do sistema')
plt.grid(True)
plt.show()
```
